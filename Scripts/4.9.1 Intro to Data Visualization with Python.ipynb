```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for mixed data types
for col in df.columns.tolist():
  weird = (df[[col]].map(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```
