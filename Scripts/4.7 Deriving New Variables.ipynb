### 3. Define function that will be used to create new column called price_label

```python
# Creating function to create price label
def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range'
    else: return 'Not enough data'
```

```python
# Apply function
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
df['prices'].max()
```

### 4. If-Statements with the loc() Function

```python
# Use loc() method on df subset to apply price range function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5) , 'price_range_loc'] = 'Mid range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Now use loc() for entire dataframe
ords_prods_merge.loc[ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High range product'
```

```python
ords_prods_merge.loc[(ords_prods_merge['prices'] <= 15) & (ords_prods_merge['prices'] > 5) , 'price_range_loc'] = 'Mid range product'
```

```python
ords_prods_merge.loc[ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low range product'
```

### 5. If-Statements with For-Loops to create new column based on busiest days

```python
ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Creating for-loop for new busiet day column
result = []

for value in ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
# Checking results of for-loop
result
```

```python
# Adding busiest day column to dataframe
ords_prods_merge['busiest_day'] = result
```

```python
ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

### 6. Create New for-loop based on busiest and slowest days

```python
#Creating new for-loop based on busiest and slowest days

new_result = []

for value in ords_prods_merge['orders_day_of_week']:
    if value == 0 or value == 1:
        new_result.append('Busiest days')
    elif value == 4 or value == 3:
        new_result.append('Slowest days')
    else:
        new_result.append('Regularly busy')
```

```python
new_result
```

```python
# Adding busiest days for-loop to datafram
ords_prods_merge['busiest_days'] = new_result
```

```python
ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

##### The total value counts for the ords_prods_merge dataframe is equal to the total sum of all groupings (i.e. "Regularly busy" + "Busiest days" + "Slowest days"). In addition, the listed sums for each label match the sum totals for the qualifying days of the week (e.g. Total "Busiest days" = 11864412, which matches orders_day_of_week 0 + 1, or 6204182 + 5660230).

### 7. Create new column "busiest_period_of_day" to identify time periods “Most orders,” “Average orders,” and “Fewest orders.”

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Create for-loop to identify time periods that are busiest

result_2 = []

for value in ords_prods_merge['order_hour_of_day']:
    if value in [10,11,14,15,13,12,16,9]:
        result_2.append('Most orders')
    elif value in [17,8,18,19,20,7,21,22]:
        result_2.append('Average orders')
    else:
        result_2.append('Fewest orders')
```

```python
# Adding busiest period of day column
ords_prods_merge['busiest_period_of_day'] = result_2
```

```python
# Print freqeuncy of busies_period_of_day column
ords_prods_merge['busiest_period_of_day'].value_counts(dropna = False)
```
